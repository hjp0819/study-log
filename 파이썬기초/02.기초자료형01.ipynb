## 문제 7) 논리 연산자 & 단락 평가
- `and`, `or`, `not` 을 사용해 조건을 완성하세요.
- 단락 평가로 인해 불필요한 `print()`가 실행되지 않도록 하세요.


```python
age = 25; has_license = True; is_student = False

cond_and = (age >= 18 ___ has_license)
cond_or  = (age >= 65 ___ is_student)
cond_not = (___ is_student)

# 단락 평가 확인 (아래 두 줄은 출력이 없어야 통과)
False ___ print("이건 실행되면 안 됨")
True ___ print("이건 실행되면 안 됨")

assert cond_and is True and cond_or is False and cond_not is True
print("✅ 문제 7 통과")
```
