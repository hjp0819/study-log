### **문제 설명**

한 주(7일)의 체온 기록을 **리스트**에 저장하고, 평균, 최고, 최저 체온을 출력하세요.

```python
temps = []

count = 0
while count < 7:
    temp = float(input(f"{count+1}번째 날 체온 입력 : "))
    temps.append(temp)
    count += 1

avg = sum(temps) / len(temps)
max_temp = max(temps)
min_temp = min(temps)

print(f"평균 체온: {avg:.2f}")
print(f"최고 체온: {max_temp:.2f}")
print(f"최저 체온: {min_temp:.2f}")
```

```python
temps = []

count = 0
while count < 7:
    temp = float(input(f"{count+1}번째 날 체온 입력 : "))
    if 36.0 <= temp <=37.5 :
        temps.append(temp)
    count += 1

print(f"정상 체온 : {temps}")
```

```python
# 미열 37.5~38.0
temps = []

count = 0
mild_fever = 0
max_days = 0

while count < 7:
    temp = float(input(f"{count+1}번째 날 체온 입력 : "))
    if 37.5 <= temp <=38.0 :
        mild_fever += 1
        max_days = max(max_days, mild_fever)
    else:
        mild_fever = 0    
    count += 1

if max_days >= 3:
    print(f"{max_days}일 이상 연속 미열입니다.")
else:
    print("연속 3일 이상 미열은 없습니다.")
```

```python
bp_dict = {}
days = ["월","화","수","목","금","토","일"]

for day in days:
    pressure = int(input(f"{day}요일 혈압 입력 : "))
    bp_dict[day] = pressure

day_input = input(f"원하는 요일을 입력하세요 : ")

key = day_input[:1]
print(f"{key}요일 혈압 : {bp_dict[key]}")
```

```python
bp_dict = {}
days = ["월","화","수","목","금","토","일"]

for day in days:
    pressure = int(input(f"{day}요일 혈압 입력 : "))
    bp_dict[day] = pressure

weekday_key = ["월","화","수","목","금"]
weekend_key = ["토","일"]

weekday_avg = sum(bp_dict[d] for d in weekday_key)/len(weekday_key)
weekend_avg = sum(bp_dict[d] for d in weekend_key)/len(weekend_key)

print(f"평일 평균 혈압: {weekday_avg:.2f}")
print(f"주말 평균 혈압: {weekend_avg:.2f}")
```
