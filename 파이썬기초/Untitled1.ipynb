```python
def log_decorator(func):
    def wrapper():
        print("[LOG] 함수 시작")
        func()
        print("[LOG} 함수 끝")
    return wrapper

#@log_decorator
@time_decorator
def greet2():
    for i in range(10000):
        pass
    print("안녕하세요!")
```

```python
def time_decorator(func):
    def wrapper():
        start = time.time()
        func()
        print('함수실행시간:',time.time()-start)
    return wrapper
```

```python
class Character:
    server = '크로아'
    def __init__(self, name, STR, LUK, DEX, INT):
        self.name = name
        self.STR = STR
        self.LUK = LUK
        self.DEX = DEX
        self.INT = INT
        self.level = 1
        self.hp = 100
        self.mp = 100

    def info(self):
        print(f"{self.name}님의 레벨은 {self.level}입니다.{self.server}")

    def attack(self):
        print(f'{self.name}이 공격했다.')
```

```python
user1 = Character('user1',6,4,7,4)
user1.info()
print(user1.server)
user1.attack()
```
