## 문제 1) 체온 평균 계산기

- 여러 환자의 체온을 받아 평균을 계산합니다.
- 옵션 `unit` 값이 `"F"`면 화씨(Fahrenheit)로 변환합니다.
- 아무 데이터도 없으면 `None`을 반환합니다.
- 결과는 소수 첫째자리까지 반올림합니다.


```python
# TODO: avg_temp(*temps, **options) 구현
def avg_temp(*temps, **options):
    """
    환자들의 체온 평균을 계산
    - *temps: 여러 체온 값
    - **options:
        - unit: "C"(섭씨, 기본값) 또는 "F"(화씨 변환)
    """
    # TODO: temps가 비어있으면 None 반환
    if not temps : 
        return None

    # TODO: 평균 계산 (총합 / 개수)
    avg = sum(temps) / len(temps)

     # TODO: unit이 "F"이면 화씨로 변환
    if options.get("unit") =="F":
        avg = avg * 9/5 + 32
    # TODO: 소수 첫째 자리까지 반올림하여 반환

    return round(avg, 1)

assert avg_temp(36.5, 37.2, 38.0) == 37.2
assert avg_temp(36.5, 37.2, 38.0, unit="F") == 99
assert avg_temp() is None
print("✅ 문제 1 통과")
```

## 문제 2) 검사 비용 계산기

- 여러 검사 항목 비용을 모두 더합니다.
- 옵션 `insurance=True`이면 절반 가격 처리합니다.
- 옵션 `discount=0.1`이면 10% 추가 할인 처리합니다.
- 최종 결제 금액을 **정수(int)** 로 반환합니다.


```python
# TODO: exam_cost(*costs, **options) 구현
def exam_cost(*costs, **options):
    """
    검사 비용 계산기
    - *costs: 여러 검사 항목 비용
    - **options:
        - insurance=True → 절반 가격
        - discount=0.1   → 추가 10% 할인
    """
    # TODO: 모든 비용 합산

    total = sum(costs)

    # TODO: 보험 적용 시 절반 가격 처리
    if options.get("insurance") == True:
        total = total*1/2

    # TODO: discount 옵션 적용
    if options.get("discount") :
        total = total * (1-float(options.get("discount")))

    # TODO: 최종 결제 금액을 int로 반환
    return int(total)

assert exam_cost(10000, 20000, insurance=True) == 15000
assert exam_cost(5000, 5000, discount=0.2) == 8000
assert exam_cost(10000, 20000, insurance=True, discount=0.1) == 13500
print("✅ 문제 2 통과")
```
