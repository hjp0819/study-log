실습: 환자 정보 관리 콘솔 
1. 환자 정보 입력 → dict 만들어 리스트에 추가 
2. 환자 목록 보기 → for와 enumerate(start=1)로 순번과 함께 출력 
3. 이름으로 검색 (부분 일치) → for + 조건 
    4. 환자 정보 수정 (메모만 간단히) → 선택한 환자의 memo 업데이트 
5. 환자 삭제 → 선택한 번호의 환자 제거 
6. 간단 통계 총 환자 수 (len) 평균 나이 (sum / len, 환자 없으면 0) 성별 분포 (count) 
7. 데이터 초기화 
8. 종료 → break


```python
patients = []


def add_pat():
    print("1. 환자 정보 입력")
    name = input("환자 이름 : ").strip()
    if not name:
        print("이름은 필수입니다."); return
    try:
        age = int(input("환자 나이 : ").strip())
    except ValueError:
        print("나이는 숫자여야 합니다."); return

    gender = input("환자 성별 (F/M): ").strip().upper()
    if gender not in ("F", "M"):
        print("성별은 F 또는 M만 입력 가능합니다."); return

    memo = input("환자 메모 : ").strip()
    patients.append({"name": name, "age": age, "gender": gender, "memo": memo})
    print("환자등록완료")


def list_pat(plist=None):
    print("2. 환자 목록 보기")
    plist = patients if plist is None else plist
    if not plist:
        print("등록된 환자가 없습니다.")
        return
    for i, p in enumerate(plist, start=1):
        print(f"{i:>2}. {p['name']} ({p['age']}세, {p['gender']}) - {p['memo']}")


def search_pat():
    print("3. 이름으로 검색 (부분 일치, 대소문자 무시)")
    if not patients:
        print("등록된 환자가 없습니다.")
        return

    key_name = input("검색할 이름: ").strip().lower()
    result = [p for p in patients if key_name in p["name"].lower()]
    if not result:
        print("검색 결과가 없습니다.")
    else:
        list_pat(result)  # 결과 출력


def mod_pat():
    print("4. 환자 정보 수정(메모만)")
    if not patients:
        print("등록된 환자가 없습니다.")
        return

    key_name = input("수정할 이름(부분 OK): ").strip().lower()
    cands = [p for p in patients if key_name in p["name"].lower()]
    if not cands:
        print("검색 결과가 없습니다."); return

    target = cands[0]   # 첫 번째만 선택
    print(f"선택: {target['name']} ({target['age']}세, {target['gender']})")
    new_memo = input("새 메모(빈칸=변경없음): ").strip()
    if new_memo:
        target["memo"] = new_memo
        print("수정 완료.")
    else:
        print("변경 없음.")


def del_pat():
    print("5. 환자 삭제")
    if not patients:
        print("등록된 환자가 없습니다.")
        return

    key_name = input("삭제할 이름(부분 OK): ").strip().lower()
    cands = [p for p in patients if key_name in p["name"].lower()]
    if not cands:
        print("검색 결과가 없습니다."); return

    target = cands[0]   # 첫 번째만 삭제
    ok = input(f"정말 삭제할까요? [{target['name']}] (Y/N): ").strip().upper()
    if ok == "Y":
        patients.remove(target)
        print("삭제 완료.")
    else:
        print("취소했습니다.")


def math_pat():
    print("6. 간단 통계")
    total = len(patients)
    avg_age = (sum(p["age"] for p in patients) / total) if total else 0
    m = sum(1 for p in patients if p["gender"] == "M")
    f = sum(1 for p in patients if p["gender"] == "F")
    print(f"총 환자 수 : {total}")
    print(f"평균 나이 : {avg_age:.1f}")
    print(f"성별 분포 : 남 {m}명, 여 {f}명")


def clear_pat():
    print("7. 데이터 초기화")
    if input("정말 초기화? (Y/N): ").strip().upper() == "Y":
        patients.clear()
        print("모두 삭제되었습니다.")
    else:
        print("취소했습니다.")


while True:
    print("""
    1. 환자 정보 입력
    2. 환자 목록 보기
    3. 이름으로 검색
    4. 환자 정보 수정
    5. 환자 삭제
    6. 간단 통계
    7. 데이터 초기화 
    8. 종료
    """)
    cmd_str = input("메뉴 번호 입력: ").strip()
    if not cmd_str.isdigit():
        print("숫자를 입력하세요.")
        continue
    cmd = int(cmd_str)

    if cmd == 1:
        add_pat()
    elif cmd == 2:
        list_pat()
    elif cmd == 3:
        search_pat()
    elif cmd == 4:
        mod_pat()
    elif cmd == 5:
        del_pat()
    elif cmd == 6:
        math_pat()
    elif cmd == 7:
        clear_pat()
    elif cmd == 8:
        print("프로그램 종료")
        break
    else:
        print("잘못된 입력입니다.")
```

```python
patients = []

# ---------- 헬퍼 ----------
def find_first(name_sub: str):
    """이름 부분일치로 첫 번째 환자의 인덱스를 반환. 없으면 None."""
    key = name_sub.lower()
    for i, p in enumerate(patients):
        if key in p["name"].lower():
            return i
    return None

def read_int(prompt: str):
    """정수 입력. 공백/문자/음수 방지. 실패 시 None."""
    s = input(prompt).strip()
    if not s.isdigit():
        return None
    v = int(s)
    if v < 0:
        return None
    return v

# ---------- 기능 ----------
def add_pat():
    print("1. 환자 정보 입력")
    name = input("환자 이름 : ").strip()
    if not name:
        print("이름은 필수입니다."); return

    age = read_int("환자 나이 : ")
    if age is None:
        print("나이는 0 이상의 숫자여야 합니다."); return

    gender = input("환자 성별 (F/M): ").strip().upper()
    if gender not in ("F", "M"):
        print("성별은 F 또는 M만 입력 가능합니다."); return

    memo = input("환자 메모 : ").strip()
    patients.append({"name": name, "age": age, "gender": gender, "memo": memo})
    print("환자등록완료")

def list_pat(plist=None, sort_key="name"):
    print("2. 환자 목록 보기")
    plist = patients if plist is None else plist
    if not plist:
        print("등록된 환자가 없습니다.")
        return
    # 정렬 옵션: 이름/나이
    if sort_key == "age":
        view = sorted(plist, key=lambda p: (p["age"], p["name"]))
    else:
        view = sorted(plist, key=lambda p: (p["name"], p["age"]))

    for i, p in enumerate(view, start=1):
        print(f"{i:>2}. {p['name']} ({p['age']}세, {p['gender']}) - {p['memo']}")

def search_pat():
    print("3. 이름으로 검색 (부분 일치)")
    if not patients:
        print("등록된 환자가 없습니다."); return
    key = input("검색할 이름: ").strip()
    if not key:
        print("입력값이 비었습니다."); return
    result = [p for p in patients if key.lower() in p["name"].lower()]
    if not result:
        print("검색 결과가 없습니다.")
    else:
        list_pat(result)

def mod_pat():
    print("4. 환자 정보 수정(메모만)")
    if not patients:
        print("등록된 환자가 없습니다."); return
    key = input("수정할 이름(부분 OK): ").strip()
    if not key:
        print("입력값이 비었습니다."); return
    idx = find_first(key)
    if idx is None:
        print("검색 결과가 없습니다."); return

    target = patients[idx]
    print(f"선택: {target['name']} ({target['age']}세, {target['gender']})")
    new_memo = input("새 메모(빈칸=변경없음): ").strip()
    if new_memo:
        target["memo"] = new_memo
        print("수정 완료.")
    else:
        print("변경 없음.")

def del_pat():
    print("5. 환자 삭제")
    if not patients:
        print("등록된 환자가 없습니다."); return
    key = input("삭제할 이름(부분 OK): ").strip()
    if not key:
        print("입력값이 비었습니다."); return
    idx = find_first(key)
    if idx is None:
        print("검색 결과가 없습니다."); return

    target = patients[idx]
    ok = input(f"정말 삭제할까요? [{target['name']}] (Y/N): ").strip().upper()
    if ok == "Y":
        patients.pop(idx)
        print("삭제 완료.")
    else:
        print("취소했습니다.")

def math_pat():
    print("6. 간단 통계")
    total = len(patients)
    avg_age = (sum(p["age"] for p in patients) / total) if total else 0
    m = sum(1 for p in patients if p["gender"] == "M")
    f = sum(1 for p in patients if p["gender"] == "F")
    print(f"총 환자 수 : {total}")
    print(f"평균 나이 : {avg_age:.1f}")
    print(f"성별 분포 : 남 {m}명, 여 {f}명")

def clear_pat():
    print("7. 데이터 초기화")
    if input("정말 초기화? (Y/N): ").strip().upper() == "Y":
        patients.clear()
        print("모두 삭제되었습니다.")
    else:
        print("취소했습니다.")

# ---------- 메인 루프 ----------
def main():
    while True:
        print("""
    1. 환자 정보 입력
    2. 환자 목록 보기
    3. 이름으로 검색
    4. 환자 정보 수정
    5. 환자 삭제
    6. 간단 통계
    7. 데이터 초기화 
    8. 종료
        """)
        try:
            cmd_str = input("메뉴 번호 입력: ").strip()
        except (KeyboardInterrupt, EOFError):
            print("\n입력이 취소되어 프로그램을 종료합니다.")
            break

        if not cmd_str.isdigit():
            print("숫자를 입력하세요."); continue
        cmd = int(cmd_str)

        if cmd == 1: add_pat()
        elif cmd == 2:
            sort_opt = input("정렬 기준 (name/age, 기본 name): ").strip().lower() or "name"
            list_pat(sort_key=sort_opt if sort_opt in ("name","age") else "name")
        elif cmd == 3: search_pat()
        elif cmd == 4: mod_pat()
        elif cmd == 5: del_pat()
        elif cmd == 6: math_pat()
        elif cmd == 7: clear_pat()
        elif cmd == 8:
            print("프로그램 종료"); break
        else:
            print("잘못된 입력입니다.")

if __name__ == "__main__":
    main()
```
