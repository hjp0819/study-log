```python
from sklearn.preprocessing import KBinsDiscretizer
import numpy as np
import pandas as pd

ages = np.array([22, 25, 35, 45, 52, 58, 67, 72]).reshape(-1, 1)

kbd = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
print(kbd.fit_transform(ages).ravel())

df = pd.DataFrame({'age': ages.ravel()})
df['age_group'] = pd.cut(df['age'], bins=[0, 30, 60, 100], labels=['청년', '중년', '노년'])
print(df)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 7, 16, 29, 46])  # y = 2x² - 3x + 2

model_linear = LinearRegression()
model_linear.fit(X, y)
print("선형 R²:", model_linear.score(X, y))

poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model_poly = LinearRegression()
model_poly.fit(X_poly, y)
print("다항식 R²:", model_poly.score(X_poly, y))
```

```python
import numpy as np
from scipy import stats

np.random.seed(42)
data = np.random.normal(100, 15, 100)
data = np.append(data, [200, 10, 250])  # 이상치 추가

z_scores = np.abs(stats.zscore(data))
outliers = np.where(z_scores > 3)
print("이상치 값:", data[outliers])
```
