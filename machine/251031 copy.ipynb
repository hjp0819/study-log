```python
import numpy as np
import matplotlib.pyplot as plt

# ① 폰트 설정 (Windows 기본 폰트: 맑은 고딕)
plt.rcParams.update({
    "font.family": "Malgun Gothic",  # 윈도우 기본 한글 폰트
    "font.size": 10,
    "axes.unicode_minus": False,     # 마이너스 깨짐 방지
})

weights = np.random.normal(loc=70, scale=8, size=100)

mean = np.mean(weights)
median = np.median(weights)
std = np.std(weights)

print(f"평균: {mean:.2f}kg, 중앙값: {median:.2f}kg, 표준편차: {std:.2f}kg")

plt.hist(weights, bins=10, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='--', label=f'평균 {mean:.1f}kg')
plt.axvline(median, color='green', linestyle=':', label=f'중앙값 {median:.1f}kg')
plt.title('체중 데이터 분포 (히스토그램)')
plt.xlabel('체중(kg)')
plt.ylabel('빈도수')
plt.legend()
plt.show()
```

```python

import numpy as np
import pandas as pd

data = pd.Series([68, 72, np.nan, 65, 80, np.nan, 75])
print("원본 데이터:")
print(data)

mean_value = data.mean()
median_value = data.median()

print("\n평균으로 채운 값:", round(mean_value, 1))
print("중앙값으로 채운 값:", round(median_value, 1))
```

```python
# 평균으로 채우기
data_mean_filled = data.fillna(mean_value)

# 중앙값으로 채우기
data_median_filled = data.fillna(median_value)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(data_mean_filled, bins=5, color='skyblue', edgecolor='black', alpha=0.6, label='평균으로 채운 데이터')
plt.hist(data_median_filled, bins=5, color='orange', edgecolor='black', alpha=0.5, label='중앙값으로 채운 데이터')
plt.legend()
plt.title('결측치 채우기 비교')
plt.xlabel('체중(kg)')
plt.ylabel('빈도수')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 극단값 포함된 데이터
data = pd.Series([68, 72, np.nan, 65, 80, np.nan, 120])  # 마지막 120이 이상치 역할
print("원본 데이터:")
print(data)

mean_value = data.mean()
median_value = data.median()

print("\n평균으로 채운 값:", round(mean_value, 1))
print("중앙값으로 채운 값:", round(median_value, 1))

# 평균과 중앙값으로 결측치 채우기
data_mean_filled = data.fillna(mean_value)
data_median_filled = data.fillna(median_value)

```

```python
# 그래프 비교
plt.figure(figsize=(6,4))
plt.hist(data_mean_filled, bins=6, color='skyblue', edgecolor='black', alpha=0.6, label='평균으로 채운 데이터')
plt.hist(data_median_filled, bins=6, color='orange', edgecolor='black', alpha=0.5, label='중앙값으로 채운 데이터')
plt.legend()
plt.title('결측치 채우기 방식 비교 (극단값 포함)')
plt.xlabel('체중(kg)')
plt.ylabel('빈도수')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(data_mean_filled, bins=np.arange(60, 130, 10), 
        color='skyblue', edgecolor='black', alpha=0.6, label='평균으로 채운 데이터')
plt.hist(data_median_filled, bins=np.arange(60, 130, 10), 
        color='orange', edgecolor='black', alpha=0.4, label='중앙값으로 채운 데이터')
plt.axvline(data_mean_filled.mean(), color='blue', linestyle='--', label='평균값')
plt.axvline(data_median_filled.mean(), color='darkorange', linestyle='--', label='중앙값')
plt.legend()
plt.title('결측치 채우기 방식 비교 (극단값 포함)')
plt.xlabel('체중(kg)')
plt.ylabel('빈도수')
plt.show()
```
