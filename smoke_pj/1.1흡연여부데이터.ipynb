```python
# IDì²˜ëŸ¼ ë¬¸ìì—´ì€ ì œì™¸í•˜ê³  ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ë§Œ
numeric_cols = health_data.select_dtypes(include=[np.number]).columns

# ê° ì»¬ëŸ¼ë³„ ìµœì†Œê°’ ì¶œë ¥
health_data[numeric_cols].min().sort_values()
```

```python
cols_zero_as_na = ['ì‹œë ¥','í˜ˆì••', 'ê³µë³µ í˜ˆë‹¹', 'ì¤‘ì„± ì§€ë°©']

for col in cols_zero_as_na:
    # ê° (AgeBand, BMIBand) ê·¸ë£¹ì˜ ì¤‘ì•™ê°’ìœ¼ë¡œ ëŒ€ì²´
    health_data[col] = (
        health_data.groupby(['AgeBand', 'BMIBand'], observed=True)[col]
        .transform(lambda x: x.fillna(x.median()))
    )
    # í˜¹ì‹œ ê·¸ë£¹ ë‚´ ì¤‘ì•™ê°’ì´ NaNì´ë©´ ì „ì²´ ì¤‘ì•™ê°’ìœ¼ë¡œ í•œ ë²ˆ ë” ëŒ€ì²´
    health_data[col] = health_data[col].fillna(health_data[col].median())

```

```python
import numpy as np

def suggest_transform(x):
    if x.min() < 0:
        return "Yeo-Johnson"
    elif x.skew() > 1:
        return "Log"
    elif 0.5 < x.skew() <= 1:
        return "Sqrt"
    elif x.skew() < -1:
        return "Square"
    else:
        return "None"

for col in health_data.select_dtypes(include=np.number).columns:
    print(f"{col:15s} â†’ {suggest_transform(health_data[col])}")
```
